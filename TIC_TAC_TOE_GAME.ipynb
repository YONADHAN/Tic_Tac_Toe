```python
#This is the TIC TAC TOE game's code
#code written by: Yonadhan M.M
#written on 06-09-2023


print("Welcome to TIC TAC TOE ")
flag = 'wrong'
# initialise ongoing = true for running the game, because on the while loop, we have given a condition 'while ongoing' so the code only runs when ongoing is true
ongoing = True
#game_limit is a list of numbers ranging from 0 to 8, it shows the position of each column to understand
game_limit = ['0','1','2','3','4','5','6','7','8']

#----------------------------Various functions required to run the TIC TAC TOE game-----------------------------------------------------------

#game_continue is a function which allows the user to continue or discontinue the game when anyone wins
def game_continue():
    global ongoing
    choice = 'wrong'
    while choice not in ['Y','N']:
        print("Type Y for continuing the game or \nType N to stop the game")
        choice = input("")
        choice = choice.upper()
        if choice == 'N':
             ongoing = False
             print("You have successfully exited")
        else:
             ongoing = True
             print("You are continuing Tic Tac Toe")
   

#This function displays the current status of the player's moves. Through this, the user can identify the positions that he can choose to put X or O
def display_status():
    print(game_limit[6]+' | '+game_limit[7]+' | '+game_limit[8])
    print('- | - | -')
    print(game_limit[3]+' | '+game_limit[4]+' | '+game_limit[5])
    print('- | - | -')
    print(game_limit[0]+' | '+game_limit[1]+' | '+game_limit[2])
    print('\n'*4)

#This is the function to ask the user to get the position input, When he types the position correctly, the position_input function returns the position value 
#place where it is called,; it will ask to enter a valid number which I can understand
def position_input():
    position = '1000'
    while position not in ['0','1','2','3','4','5','6','7','8']:
        position = input("Enter positions within 0 to 8 ")
        if position not in ['0','1','2','3','4','5','6','7','8']:
            print("Enter a valid number that i can understand")
    position = int(position)        
    return position        

#This is the function to replace the game_limit elements with X and O as per the rules of TIC TAC TOE game
def displacement(position,flag):
    if flag == 'X':
        game_limit[position] = 'O'
        flag = 'O'
    else:
        game_limit[position] = 'X'
        flag = 'X'
    return game_limit, flag
   
#This is the function to check the current status of victory
def checking_victory(game_limit):
    victorycheckbox = [[0,1,2],[3,4,5],[6,7,8],[6,3,0],[7,4,1],[8,5,2],[8,4,0],[6,4,2]]
    for x, y, z in victorycheckbox:
        if game_limit[x] == game_limit[y] == game_limit[z] == flag:
            print("You won the game\n\nCongratulations {}".format(flag))
            display_status()
            game_continue()
                
 #------------------------------------Main codes-----------------------------------     
while flag not in ['X','O']:
    flag = input("Enter X or O , Who is going to play the game first")
    flag = flag.upper()
    print(flag)
    if flag not in ['X','O']:
        print("Enter X or Y only")
    if flag == 'X':
        flag = 'O'
    else:
        flag = 'X'
#This function will execute the game, whenever ongoing is true       
while ongoing:
    display_status()
    position=position_input()
    game_limit, flag = displacement(position,flag)
    checking_victory(game_limit)
    display_status()

```
